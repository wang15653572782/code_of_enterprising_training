数据读取

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#导入数据
all_data=pd.ExcelFile(r"Province\in.xlsx")
```

```python
data = []
for sheet in all_data.sheet_names:
    df = all_data.parse(sheet_name=sheet)
    data.append(df)
```

```python
#将不同的sheet拼接(axis=0向下拼接)
data_2019 = pd.concat(data,axis=0)
data_2019.head()
```

```python
data_2019=data_2019.drop(labels='城市代码', axis=1)
```

```python
data_2019
```

```python
peking=data_2019.loc[0]
```

```python
peking.index=[2019,2020,2021,2022,2023]
```

```python
peking.loc[2019]
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,41,1)
plt.plot(x,peking.loc[2019][1:],label='2019')
plt.plot(x,peking.loc[2020][1:],label='2020')
plt.plot(x,peking.loc[2021][1:],label='2021')
plt.plot(x,peking.loc[2022][1:],label='2022')
plt.plot(x,peking.loc[2023][1:],label='2023')
plt.legend()
plt.show()
```
